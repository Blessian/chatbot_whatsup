```python
from pathlib import Path
import copy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

```python
path = Path("/content/drive/MyDrive/메타버스 아카데미/본과정/Python/6. Deep_NLP/miniproj_gpt2/all_chatbot_data.csv")
data = pd.read_csv(str(path))
data.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
data
```

```python
df = copy.deepcopy(data)
```

```python
df["Q Length"] = df["Q"].apply(lambda text: len(text))
```

```python
df["A Length"] = df["A"].apply(lambda text: len(str(text)))
```
