# Glob pdf files

```python
from pathlib import Path
```

```python
pdf_path_generator = Path("./").glob("data/*.pdf")
```

# Read pdf

```python
import pandas as pd
import re
import pdfreader
from pdfreader import PDFDocument, SimplePDFViewer
```

```python
df = pd.DataFrame(columns=["Subject", "Date", "No", "Question", "1", "2", "3", "4", "Answer", "Include_Image", "Image_index", "Image_Path", "Category"])
```

```python
for pdf_path in pdf_path_generator:
    file_name = pdf_path.stem

    # subject, date
    subject = file_name[:6]
    date = file_name[6:14]

    fd = open(str(pdf_path), "rb")
    viewer = SimplePDFViewer(fd)

    page_num = 1
    all_text_list = list()
    all_image_path_list = list()
    while True:
        try:
            viewer.navigate(page_num)
        except pdfreader.PageDoesNotExist as e:
            print("End of Page")
            break
        viewer.render()

        # extract text
        text = viewer.canvas.strings
        text_of_page = "".join(text)
        all_text_list.append(text_of_page)

        # extract images
        images_dict = viewer.canvas.images
        if len(images_dict.keys()) > 0:
            image_dir = Path("./data/images").joinpath(pdf_path.stem)
            image_dir.mkdir(parents=True, exist_ok=True)

            for key, image_object in images_dict.items():
                image = image_object.to_Pillow()
                image_path = str(image_dir.joinpath(key[2:]).with_suffix(".png"))
                image.save(image_path, "png")
                all_image_path_list.append(image_path)

        page_num += 1
    print(" ".join(all_text_list))
    break
```

```python
all_text = " ".join(all_text_list)
print(all_text)
```
