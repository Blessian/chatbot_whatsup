# Token

```python
Q_TKN = "<usr>"
A_TKN = "<sys>"
BOS = '</s>'
EOS = '</s>'
MASK = '<unused0>'
SENT = '<unused1>'
PAD = '<pad>'
```

# Tokenizer

```python
from transformers import PreTrainedTokenizerFast, GPT2LMHeadModel

koGPT2_TOKENIZER = PreTrainedTokenizerFast.from_pretrained("skt/kogpt2-base-v2",
            bos_token=BOS, eos_token=EOS, unk_token='<unk>',
            pad_token=PAD, mask_token=MASK)
model = GPT2LMHeadModel.from_pretrained('skt/kogpt2-base-v2')
```

# Inference

```python
import torch
```

```python
from pathlib import Path
```

```python
import time
```

```python
dir = Path("/content/drive/MyDrive/메타버스 아카데미/본과정/Python/6. Deep_NLP/miniproj_gpt2")
```

```python
model = GPT2LMHeadModel.from_pretrained('skt/kogpt2-base-v2')
model.load_state_dict(torch.load(str(dir.joinpath("chatbot_gpt2_1.pth")), map_location=torch.device('cpu')))

sent = "0"

with torch.no_grad():
    while 1:
        q = input("user > ").strip()
        if q == "quit":
            break
        a = ""
        start_time = time.time()
        while 1:
            input_ids = torch.LongTensor(koGPT2_TOKENIZER.encode(Q_TKN + q + SENT + sent + A_TKN + a)).unsqueeze(dim=0)
            pred = model(input_ids)
            pred = pred.logits
            gen = koGPT2_TOKENIZER.convert_ids_to_tokens(torch.argmax(pred, dim=-1).squeeze().numpy().tolist())[-1]

            if gen == EOS: break
            if time.time() - start_time > 7: print("Time out!"); break

            a += gen.replace("▁", " ")
        print("Chatbot > {}".format(a.strip()))
```
